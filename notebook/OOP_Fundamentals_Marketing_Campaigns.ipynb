## Demo: BankAccount Class

```python
class BankAccount:
    def __init__(self, account_holder_name, initial_balance):
        self.account_holder_name = account_holder_name
        self.balance = initial_balance

    def deposit(self, deposit_amount):
        self.balance = self.balance + deposit_amount
        print(f"Deposit successful. Updated balance: {self.balance}")

    def withdraw(self, withdrawal_amount):
        if withdrawal_amount > self.balance:
            print("Error: Withdrawal amount exceeds available balance.")
        else:
            self.balance = self.balance - withdrawal_amount
            print(f"Withdrawal successful. Updated balance: {self.balance}")

    def show_balance(self):
        print("Current Balance:", self.balance)
```

```python
bank_account = BankAccount("Ram", 5000)
bank_account.show_balance()

bank_account.deposit(4000)
bank_account.show_balance()

bank_account.withdraw(12000)
bank_account.show_balance()
```

## Question 1: Marketing Campaign Details

```python
'''
Coding Question 1: Marketing Campaign Details

Create a class MarketingCampaign with 3 functions:

Requirements:
__init__()  - Accept campaign_name, platform, budget
apply_discount() - If budget > 50,000 → apply 10% discount; Else → apply 5% discount
show_details() - Display campaign name, platform, final budget
'''

class MarketingCampaign:
    def __init__(self, campaign_name, platform, budget):
        self.campaign_name = campaign_name
        self.platform = platform
        self.budget = budget
        self.discount_amount = 0

    def apply_discount(self):
        if self.budget > 50000:
            self.discount_amount = self.budget * 0.10
        else:
            self.discount_amount = self.budget * 0.05

    def show_details(self):
        final_budget = self.budget - self.discount_amount
        print("Campaign Name:", self.campaign_name)
        print("Platform:", self.platform)
        print("Final Budget:", final_budget)
```

```python
campaign_1 = MarketingCampaign("Survey", "Online", 60000)
campaign_1.apply_discount()
campaign_1.show_details()
```

## Question 2: Lead Cost Calculation

```python
'''
Coding Question 2: Lead Cost Calculation

Create a class LeadCostCalculator with 3 functions:

Requirements:
__init__()       - Accept campaign_name, total_leads, total_cost
cost_per_lead()  - Calculate cost per lead
display_report() - Print campaign name and cost per lead
'''

class LeadCostCalculator:
    def __init__(self, campaign_name, total_leads, total_cost):
        self.campaign_name = campaign_name
        self.total_leads = total_leads
        self.total_cost = total_cost
        self.lead_cost = 0

    def cost_per_lead(self):
        self.lead_cost = self.total_cost / self.total_leads

    def display_report(self):
        print("Campaign Name:", self.campaign_name)
        print("Cost Per Lead:", self.lead_cost)
```

```python
lead_campaign = LeadCostCalculator("Survey", 2, 500)
lead_campaign.cost_per_lead()
lead_campaign.display_report()
```

## Question 3: Campaign Performance Status

```python
'''
Coding Question 3: Campaign Performance Status

Create a class CampaignPerformance with 3 functions:

Requirements:
__init__()           - Accept campaign_name, impressions, conversions
performance_status() - If conversion rate >= 5% → "Excellent"; 2%-4.99% → "Average"; Else → "Poor"
show_status()        - Display campaign name and performance status
'''

class CampaignPerformance:
    def __init__(self, campaign_name, impressions, conversions):
        self.campaign_name = campaign_name
        self.impressions = impressions
        self.conversions = conversions
        self.status = ""

    def performance_status(self):
        self.conversion_rate = self.conversions / self.impressions  # FIX: store as instance variable

        if self.conversion_rate >= 0.05:
            self.status = "Excellent"
        elif 0.02 <= self.conversion_rate < 0.05:  # FIX: corrected operator from > to <
            self.status = "Average"
        else:
            self.status = "Poor"

    def show_status(self):
        print("Campaign Name:", self.campaign_name)
        print("Performance Status:", self.status)
```

```python
performance_campaign = CampaignPerformance("Ad Campaign", 1000, 60)
performance_campaign.performance_status()
performance_campaign.show_status()
```

## Question 5: ROI Calculation

```python
'''
Coding Question 5: ROI Calculation

Create a class ROICalculator with 3 functions:

Requirements:
__init__()      - Accept campaign_name, revenue, investment
calculate_roi() - ROI = ((revenue - investment) / investment) * 100
print_roi()     - Display campaign name and ROI value
'''

class ROICalculator:
    def __init__(self, campaign_name, revenue, investment):
        self.campaign_name = campaign_name
        self.revenue = revenue
        self.investment = investment
        self.roi = 0

    def calculate_roi(self):
        self.roi = ((self.revenue - self.investment) / self.investment) * 100

    def print_roi(self):
        print("Campaign Name:", self.campaign_name)
        print("ROI:", round(self.roi, 2), "%")
```

```python
roi_campaign = ROICalculator("Instagram Ads", 50000, 20000)
roi_campaign.calculate_roi()
roi_campaign.print_roi()
```

## Question 6: Customer Reach Category

## Question 8: Social Media Engagement

```python
'''
Coding Question 8: Social Media Engagement

Create a class SocialMediaEngagement with 3 functions:

Requirements:
__init__()         - Accept campaign_name, likes, shares, comments
calculate_engagement() - Engagement = likes + shares + comments
show_engagement()  - Display campaign name and engagement score
'''

class SocialMediaEngagement:
    def __init__(self, campaign_name, likes, shares, comments):
        self.campaign_name = campaign_name
        self.likes = likes
        self.shares = shares
        self.comments = comments
        self.engagement_score = 0  # FIX: defined as attribute, not method

    def calculate_engagement(self):  # FIX: renamed method to avoid clash with attribute
        self.engagement_score = self.likes + self.shares + self.comments

    def show_engagement(self):
        print("Campaign Name:", self.campaign_name)
        print("Engagement Score:", self.engagement_score)
```

## Question 9: Marketing Budget Utilization

```python
'''
Coding Question 9: Marketing Budget Utilization

Create a class BudgetUtilization with 3 functions:

Requirements:
__init__()       - Accept campaign_name, allocated_budget, used_budget
budget_status()  - If used_budget > allocated_budget → "Over Budget"; Else → "Within Budget"
display_budget() - Print campaign name and budget status
'''

class BudgetUtilization:
    def __init__(self, campaign_name, allocated_budget, used_budget):
        self.campaign_name = campaign_name
        self.allocated_budget = allocated_budget
        self.used_budget = used_budget
        self.status = ""

    def budget_status(self):
        if self.used_budget > self.allocated_budget:
            self.status = "Over Budget"   # FIX: corrected typo "Budegt" → "Budget"
        else:
            self.status = "Within Budget" # FIX: corrected typo "Budegt" → "Budget"

    def display_budget(self):
        print("Campaign Name:", self.campaign_name)
        print("Budget Status:", self.status)
```

```python
budget_campaign = BudgetUtilization("Instagram Ad", 200, 150)
budget_campaign.budget_status()
budget_campaign.display_budget()
```

## Question 10: Email Campaign Success Rate

```python
'''
Coding Question 10: Email Campaign Success Rate

Create a class EmailCampaign with 3 functions:

Requirements:
__init__()     - Accept campaign_name, emails_sent, emails_opened
success_rate() - Calculate open rate percentage: (emails_opened / emails_sent) * 100
show_result()  - Display campaign name and success rate
'''

class EmailCampaign:
    def __init__(self, campaign_name, emails_sent, emails_opened):
        self.campaign_name = campaign_name
        self.emails_sent = emails_sent
        self.emails_opened = emails_opened
        self.open_rate_percentage = 0

    def success_rate(self):
        self.open_rate_percentage = (self.emails_opened / self.emails_sent) * 100  # FIX: corrected formula

    def show_result(self):
        print("Campaign Name:", self.campaign_name)
        print("Success Rate:", self.open_rate_percentage, "%")
```

```python
email_campaign = EmailCampaign("Instagram Ad", 100, 70)
email_campaign.success_rate()
email_campaign.show_result()
```
